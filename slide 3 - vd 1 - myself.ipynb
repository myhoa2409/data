## lệnh truy vấn Query()

```python
#dfquery((so sánh 1 ) and(so sánh 2)
data.query("(rating_good>1000)and(rating_star>4)").filter(['name','shop_location'])
```

```python
data.query("(rating_bad<200)and(join_year<=2017)and(is_shopee_verified ==1)").filter(['name','shop_location'])
```

```python
data.query("shop_location.str.contains('Manila')==True").filter(['name','shop_location'])
```
