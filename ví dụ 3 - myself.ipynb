## SỬ DỤNG LOC VÀ ILOC

## sử dụng .loc và .iloc

```python
# Thêm 2 dòng có index -1, 2.5 vào bảng
data.loc[-1] = ['ThanhHoa', 66, 'Nem chua', 5, 'KG', 6, 2022, 66.1234]
data.loc[2.5] = ['DaNang', 77, 'BTCTH', 5, 'KG', 6, 2022, 77.1234]
data = data.sort_index()
display(data.head(10))
```

## truy xuất 1 dòng dữ liệu

```python
#df.loc[index] - dùng loc thì là vị trí đã được tính sẵn trong bảng
data.loc[2.5]
```

```python
#df.iloc[index] - iloc là vị trí đến từng dòng
data.iloc[4]
```

## truy xuất nhiều dòng rời rạc

```python
#df.loc[index bắt đầu: index kết thúc]
#df.iloc[vị trí bắt đầu: vị trí kết thúc]
data.iloc[0:5]
```
