```python
import pandas as pd
import numpy as np
print('khai báo thư viện thành công')
data=pd.read_excel('house_price_dống-da.xlsx')
display(data)
print("Đã nhập dữ liệu thành công")
```

## Yêu cầu: 3. Phát hiện các dòng, cột chứa dữ liệu khuyết thiếu Xóa bỏ hết tất cả những dòng dữ liệu không có thông tin về giá

```python
# phát hiện dòng cột chứa dự liệu khuyết thiếu
data.isna() # kiểm tra phần tử nào bị khuyết, kết quả Ture là bị khuyết thiếu
# data.notna()  # kiểm tra phần tử nào không bị khuyết, kết quả False tức là bị khuyết thiếu
data.isna().describe()

data.isnull()

data.isnull().sum()
```

```python
#Xóa bỏ hết tất cả những dòng dữ liệu không có thông tin về giá
data1 = data
# Nếu muốn xóa những hàng mà giá trị khuyết phải nằm trên các cột xác định trước ta làm như sau:
a = data1.dropna(subset=['price'])
a
```

## 4.Thực hiện xử lý giá trị khuyết thiếu: Thay thế giá trị khuyết thiếu của land_certificate bằng =”không có thông tin”, house_direction, balcony_direction, toilet, bedroom, Floor bằng giá trị có tần số xuất hiện lớn nhất của các thuộc tính đó

```python
# xử lý dự liệu khuyết thiếu
# Thay thế giá trị khuyết thiếu của land_certificate bằng =”không có thông tin”
values = {"land_certificate": 'không có thông tin', "house_direction": data1.loc[:,"house_direction"].mode(),
          "balcony_direction": data1.loc[:,"balcony_direction"].mode(), 
          "toilet": data1.loc[:,"toilet"].mode(),
          "bedroom":data1.loc[:,"bedroom"].mode(),
          "floor":data1.loc[:,"floor"].mode() } 
data1.fillna(value=values)
```

## 5. Lọc thông tin những bất động sản ở trong ngõ thành bộ dữ liệu nhà ngõ Tính toán giá/m2 ( đơn vị triệu/m2) với loại hình nhà mặt phố

## 6. Phát hiện giá trị ngoại lai của các thuộc tính: diện tích, giá/m2 bằng phương pháp IQR Thực hiện loại bỏ các dòng dữ liệu ngoại lai

```python
# Xác định dữ liệu ngoại lai bằng IQR Score
# • Tính toán IQR Score
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
# • Xác định ngoại lai
data3 = (data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))

# Xóa dữ liệu theo IQR Score:
data4 = data[~((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 *IQR))).any(axis=1)]
```
