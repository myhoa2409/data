Bộ dữ liệu FoodPrice InTurkey đã được giới thiệu từ các buổi trước. Hãy thực hiện các kiểm định sau:

Với mức ý nghĩa 5% hay kiểm định giả thuyết: giá bán lẻ gạo trung bình năm 2019 là 9.5 (Lira)/1 kg Turkish Lira là đơn vị tiền tệ ở Turkey (1usd ~ 8-9 Lira)

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
print('Khai báo thư viện thành công')
```

```python
data = pd.read_csv('FoodPrice_in_Turkey.csv', header=0)
data
```

```python
rice = data.query("ProductName == 'Rice - Retail' and Year == 2019")
rice
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(rice['Price'], bins=5)
```

```python
rice['Price'].mean()
```

```python
rice.info()
```

Với mức ý nghĩa 5% hay kiểm định giả thuyết: giá bán lẻ gạo trung bình năm 2019 là 9.5 (Lira)/1 kg

```python
stats.ttest_1samp(rice['Price'], 9.5)
```

## Ví dụ 2: Kiểm định 2 mẫu độc lập

Với mức ý nghĩa 5% hãy kiểm định giả thuyết: Giá bột mỳ và giá gạo ở Turkey năm 2019 là bằng nhau

```python
mylist = ['Wheat flour - Retail', 'Rice - Retail']
mygao = data.query("ProductName in @mylist and Year == 2019")
mygao
```

```python
my = data.query("ProductName == 'Wheat flour - Retail' and Year == 2019")
gao = data.query("ProductName == 'Rice - Retail' and Year == 2019")
display(my)
display(gao)
```

```python
stats.ttest_ind(gao['Price'], my['Price'])
```

```python
print('gia gao', gao['Price'].mean())
print('gia my', my['Price'].mean())
```
