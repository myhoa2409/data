## Đọc bộ dữ liệu, cho biết số dòng, số cột và kiểu dữ liệu của các thuộc tính.

```python
import pandas as pd
import numpy as np
data=pd.read_excel('house_price_dống-da.xlsx')
data
data.info()
data.shape[0]
data.shape[1]
```

## Lọc ra các bản ghi bán nhà riêng tại phường Trung liệt hoặc phường Khâm Thiên

```python
data
data1 = data[(data['ward_name']=='Phường Trung Liệt') | (data['ward_name']=='Phường Khâm Thiên')].filter(['title','ward_name'])
data1
```

## Lọc các thông tin Địa chỉ, Giá, Hướng nhà, Hướng ban công của các bản ghi có giấy chứng nhận sổ đỏ và có 3 phòng ngủ trở lên.

```python
data
data2 = data[(data['land_certificate']=='Sổ đỏ')|(data['bedroom']>=3)].filter(['address','price','house_direction','land_certificate','bedroom'])
data2
```

## Với mỗi loại nhà đất, tính trung bình cộng giá cũng như giá lớn nhất và giá nhỏ nhất.

```python
data
data3 = data[(data['land_certificate']=='Sổ đỏ')|(data['bedroom']>=3)].filter(['address','price','house_direction','land_certificate','bedroom'])
data3
```

## Tính trung bình cộng số phòng ngủ, số phòng vệ sinh, số tầng của mỗi phường.

```python
data4 = data.pivot_table(values=['bedroom','toilet','floor'],index='title',aggfunc ='mean')
data4
```
