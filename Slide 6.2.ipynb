```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [2.0, 2.3, 4, 3.1, 5]
y = [3, 4, 5.5, 4.7, 7.3]
```

```python
plt.scatter(x,y);
```

```python
mx = np.mean(x)
my = np.mean(y)
subx = x-mx
suby = y-my
tu = []
# for i in range(len(x)):
#     tu.append(subx[i]*suby[i])
tong_tu = np.sum(subx*suby)
tong_tu
```

```python
mau = (np.sum(subx**2)*np.sum(suby**2))**0.5
mau
```

```python
r = tong_tu/mau
r
```

```python
def correlation(x, y):
    # Finding the mean of the series x and y
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    # Subtracting mean from the individual elements
    sub_x = [i-mean_x for i in x]
    sub_y = [i-mean_y for i in y]
    # covariance for x and y
    numerator = sum([sub_x[i]*sub_y[i] for i in range(len(sub_x))])
    # Standard Deviation of x and y
    std_deviation_x = sum([sub_x[i]**2.0 for i in range(len(sub_x))])
    std_deviation_y = sum([sub_y[i]**2.0 for i in range(len(sub_y))])
    # squaring by 0.5 to find the square root
    denominator = (std_deviation_x*std_deviation_y)**0.5 
    cor = numerator/denominator
    return cor
print(correlation(x,y))
```

```python
import scipy.stats as stats

stats.pearsonr(x, y)
```

p-value nhỏ -> 2 đặc trưng có quan hệ với nhau
r dương -> 2 đặc trưng có quan hệ tỉ lệ thuận

```python
z = -2*np.array(x)
plt.scatter(x, z)
```

```python
stats.pearsonr(x, z)
```

r = -1 --> x và z có tương quan tỉ lệ nghịch (tương quan chặt)
p-value =0 --> x và z chắc chắn 100% có mối quan hệ tương quan

Kiểm định mối tương quan x và y bằng kiểm định pearson

```python
import seaborn as sns
import pandas as pd
```

```python
x = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
y = [0.353, 3.517, 0.125, -7.545, -0.555, -1.536, 3.350, -1.578, -3.537, -1.579]
data = pd.DataFrame(columns = ['x', 'y'])
data['x']=x
data['y']=y
data
```

```python
sns.lmplot(x='x', y='y', data=data)
```

```python
stats.pearsonr(x, y)
```

r = 0 --> ko có quan hệ
0.2 < |r| < 0.6 --> quan hệ lỏng
0.6 < |r|  --> quan hệ chắc chắn
