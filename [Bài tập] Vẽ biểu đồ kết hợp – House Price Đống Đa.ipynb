```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("house_price_dống-da.xlsx")
data
```

## Phân tích mối liên hệ giữa diện tích với giá nhà. Đồng thời, giữa số phòng ngủ với giá nhà và giữa số toilet với giá nhà.

```python
plt.bar(data.loc[:,'area'], data.loc[:,'price'])
plt.show()
# Tinh chỉnh thuộc tính biểu đồ
# In [45]:

plt.bar(data.loc[:,'area'], data.loc[:,'price'], width = 0.5)

plt.title('So sánh giá nhà và diện tích', fontsize = 16, color = 'r')
plt.xlabel('area', fontsize = 5)
plt.ylabel('price', fontsize = 14)
plt.show()
```
