## Thêm 1 dòng mới vào cuối bảng
### df.loc[cs]=[....]
df.loc[df.shape[0]] = [....]

```python
# Sai lầm
# data.loc[data.shape[0]]= ['Sài Gòn','3000','Dưa hấu','5','KG','10','6','2022','50.000','0','NaN']
# Chuẩn
data.loc[data.shape[0]]= ['Sài Gòn',3000,'Dưa hấu',5,'KG',10, 6,2022,50.000,0,np.NaN]
display (data)
data.info()
```

### Thêm 1 dòng mới vào đầu bảng

```python
# Thêm 1 dòng mới vào đầu bảng
data.loc[-1]= ['Hà Nội',3001,'Cốm làng Vòng',5,'KG',10, 6,2022,50.000,0,np.NaN]
# Update index của bảng
data.index = data.index + 1
# Sắp xếp lại bảng
data = data.sort_index()
display(data)
data.info()
```

### Thêm 1 dòng mới vào vị trí bất kỳ

```python
# Chèn thêm dòng ở vị trí bất kỳ, ví dụ 3
# data.loc[2.5]= ['Huế',3002,'Cơm hến',5,'KG',10, 6,2022,50.000,0,np.NaN]

#Sắp xếp lại các hàng theo đúng thứ tự từ nhỏ tới lớn và theo số tự nhiên
data = data.sort_index().reset_index(drop=True)

display(data)
```

## Xóa cột

## Xóa cột sử dụng lệnh del
### del df['Tên cột cần xóa']

```python
# Xóa cột sử dụng del
data2 = data
display(data2)
# Xóa cột 'cột mới' (xóa ngay trên bảng gốc) -> Nguy hiểm, hạn chế dùng
del data2['cột mới']
display(data2)
display(data)
```

## Xóa cột sử dụng hàm pop()
### df.pop('Tên cột cần xóa')

```python
# Xóa cột sử dụng pop()
data3 = data
display(data3)
# Xóa cột 'cột mới' (xóa ngay trên bảng gốc) -> hạn chế dùng
data3.pop('cột mới') #-> kết quả là cột cần xóa
display(data3)
display(data)
```

## Xóa cột sử dụng hàm drop()
### df.drop('Tên cột cần xóa’, axis=1, inplace=True/False)
axis = 1 nghĩa là xóa cột
inplace = True -> xóa cả dữ liệu gốc = del, pop()
inplace = False -> dữ lại nguyên vẹn dữ liệu gốc (mặc định)

```python
# Xóa cột sử dụng drop()
data4 = data
print('Data4')
display(data4)
# Xóa cột 'cột mới' (không xóa trực tiếp trên bảng gốc) -> nên dùng
data5 = data4.drop('cột mới', axis=1)  #-> lấy kết quả trả về
print('Data4')
display(data4)
print('Data5')
display(data5)
print('Data gốc')
display(data)
```

## Xóa nhiều cột sử dụng hàm drop()
### df.drop(['Tên cột cần xóa1',…, 'Tên cột cần xóa n'], axis=1, inplace=True/False)

```python
# Xóa cột 'cột mới' và cột 'số lượng'
data6 = data.drop(['cột mới', 'Số lượng'], axis=1)
print('Data6')
display(data6)
print('Data gốc')
display(data)
```

## Xóa dòng sử dụng hàm drop()
### df.drop(cs, axis=0, inplace=True/False)
axis = 0 -> xóa dòng
cs là index của dòng đấy

```python
#Xóa dòng cuối cùng
data6 = data.drop(data.shape[0]-1, axis=0)
print("Data6")
display(data6)
```
