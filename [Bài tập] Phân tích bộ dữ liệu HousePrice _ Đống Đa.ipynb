## [Bài tập] Phân tích bộ dữ liệu HousePrice _ Đống Đa

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import ipywidgets as widgets
print('Khai báo thư viện thành công')
#Nhập dữ liệu
data = pd.read_excel('house_price_dống-da.xlsx', header=0)
```

phân tích mối liên hệ giữa diện tích với giá nhà## Vẽ biểu đồ phân tích mối liên hệ giữa diện tích với giá nhà, giữa số phòng ngủ với giá nhà, giữa số toilet với giá nhà.

```python
list_ = ['price','area','bedroom','toilet']
data1 = data.loc[:,['price','area','bedroom','toilet']]
data1.isna().sum()
data2 = data1.dropna(subset='price')
data3 = data2.fillna(0)
data3
```

```python
plt.scatter(data3.loc[:,'area'], data3.loc[:,'price'])
plt.title('phân tích mối liên hệ giữa diện tích với giá nhà', fontsize = 16, color = 'r')
plt.xlabel('giá', fontsize = 14)
plt.ylabel('diện tích', fontsize = 14)
plt.show()
```

```python
plt.scatter(data3.loc[:,'price'], data3.loc[:,'bedroom'])
plt.title('phân tích mối liên hệ giữa diện tích với giá nhà', fontsize = 16, color = 'r')
plt.xlabel('giá', fontsize = 14)
plt.ylabel('bedroom', fontsize = 14)
plt.show()
```

```python
plt.scatter(data3.loc[:,'price'], data3.loc[:,'toilet'])
plt.title('phân tích mối liên hệ giữa diện tích với giá nhà', fontsize = 16, color = 'r')
plt.xlabel('giá', fontsize = 14)
plt.ylabel('toilet', fontsize = 14)
plt.show()
```

## Vẽ biểu đồ so sánh giá nhà trung bình trên 1 m2 giữa các hình thức nhà (type_of_land).

```python
data4 =data.loc[:,['type_of_land','area','price']]
data4['giá nhà tb']=data4.loc[:,'price']/data4.loc[:,'area']
data5 = data4.dropna(subset='giá nhà tb')  
data5
```

```python
plt.bar(data5['type_of_land'], data5['giá nhà tb'],width = 0.5)
# Tinh chỉnh thuộc tính biểu đồ

plt.title('giá nhà trung bình trên 1 m2 giữa các hình thức nhà (type_of_land)', fontsize = 16, color = 'r')
plt.xlabel('type_of_land', fontsize = 14)
plt.ylabel('Price', fontsize = 14)
plt.show()
```

## Vẽ biểu đồ thể hiện tỉ lệ % bài đăng (bản ghi) giữa các hình thức nhà (type_of_land).

```python
#Đánh giá tỷ lệ nên dùng biểu đồ tròn
plt.pie(data5.iloc[:,1], labels = data5.iloc[:,0])
plt.show()
```
