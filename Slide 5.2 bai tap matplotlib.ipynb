```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import ipywidgets as widgets
print('Khai báo thư viện thành công')
#Nhập dữ liệu
data = pd.read_csv('granted-citizenship-1949-2019.csv', header=0, encoding= 'ISO-8859-1')
data
```

```python
data.info()
```

```python
country = data.loc[0:323,'Country of Birth']
country
```

```python
comboBox = widgets.Combobox(
    placeholder='Quốc gia',
    options=country.tolist(),
    description='Lựa chọn quốc gia:',
    ensure_option=True,
    disabled=False,
    value='Vietnam',
    continuous_update = True
)
display(comboBox)
```

```python
SliderYear = widgets.IntSlider(
    description='Năm',
    min=1949, 
    max=2019, 
    step=1, 
    value=1990,
    orientation='horizontal',
    continuous_update = False)
display(SliderYear)
```

```python
data = data.fillna(0)
display(data)
```

```python
from ipywidgets import interact as interact

# @interact(country_=comboBox, year_=SliderYear)
def my_fig(country_, year_):
    fig1 = plt.figure()
    temp = data.loc[data['Country of Birth']==country_,'1949':'2019'] #Lọc ra nước cần tìm
    plt.plot(np.arange(1949,2020), temp.iloc[0,:]) #Vẽ biểu đồ đường
    ax = plt.gca()
    ax.set_title(country_)
    ax.set_xlabel('Năm')
    ax.set_ylabel('Số lượng người nhập cư')
    plt.scatter(year_, temp.loc[:,str(year_)],c='r') #Vẽ biểu đồ điểm
    text_ = temp.loc[:,str(year_)].item() #Hiển thị giá trị của điểm 
    space = 0.05*(temp.max(axis=1) - temp.min(axis=1))
    # print(data.loc[data['Country of Birth']==country_,'1949':'2019'].max(axis=1))
    plt.annotate(text_,(int(year_)-1, temp.loc[:,str(year_)]+space), rotation=90)
    plt.show()

interact(my_fig, country_=comboBox, year_=SliderYear);
```

```python
DNA = ['Vietnam', 'Thailand', 'Cambodia', 'Laos', 'Malaysia', 'Philippines','Indonesia',
      'Myanmar', 'Singapore', 'East Timor', 'Brunei']
```

```python
comboBoxDNA = widgets.Combobox(
    placeholder='Quốc gia',
    options=DNA,
    description='Lựa chọn quốc gia:',
    ensure_option=True,
    disabled=False,
    value='Vietnam',
    continuous_update = True
)
display(comboBoxDNA)
```

```python
data_DNA = data.query('`Country of Birth` in @DNA').drop(columns=['Total', '%'])
data_DNA
```

```python
def my_fig_DNA(country_, year_):
    fig2, ax = plt.subplots(2,2)
    temp = data_DNA.loc[data['Country of Birth']==country_,'1949':'2019'] #Lọc ra nước cần tìm
    ax[0][0].plot(np.arange(1949,2020), temp.iloc[0,:]) #Vẽ biểu đồ đường
    ax[0][0].set_title(country_)
    ax[0][0].set_xlabel('Năm')
    ax[0][0].set_ylabel('Số lượng người nhập cư')
    ax[0][0].scatter(year_, temp.loc[:,str(year_)],c='r') #Vẽ biểu đồ điểm
    text_ = temp.loc[:,str(year_)].item() #Hiển thị giá trị của điểm 
    space = 0.05*(temp.max(axis=1) - temp.min(axis=1))
    # print(data.loc[data['Country of Birth']==country_,'1949':'2019'].max(axis=1))
    ax[0][0].annotate(text_,(int(year_)-1, temp.loc[:,str(year_)]+space), rotation=90)
    
    ax[0][1].pie(data_DNA.loc[:,str(year_)], labels=data_DNA.loc[:,'Country of Birth'])
    
    plt.show()

interact(my_fig_DNA, country_=comboBoxDNA, year_=SliderYear);
```
