## Câu 10. Hãy tính toán lại các cột 'Tổng', '%', và hàng 'TỔNG' sao cho đúng. Hiển thị bảng.
## Thực hiện các phép tính tương tự như câu 7, hãy đánh giá xem các số liệu được thêm vào ảnh hưởng thế nào tới kết quả.

**Gợi ý:**

```python
# dùng dữ-liệu.sum(axis=0 hoặc 1) để tính tổng, trong đó 0 hoặc 1 là hướng tính tổng theo chiều dọc hoặc chiều ngang
# Chuyển đổi float sang số% và ngược lại, tham khảo https://levelup.gitconnected.com/convert-percentage-string-to-numeric-and-vice-versa-in-pandas-18a3d66e2853
```

```python
%%cauhoi 10
%%tee response
# Code dưới dòng này

data.loc[:,'Tổng']=data.loc[0:324,'1949':'2020'].sum(axis=1) #Tính toán lại cột Tổng
data.loc[325,'Tổng']=data.loc[:324,'Tổng'].sum() #Tính toán lại ô TỔNG, Tổng

sum_ = data.loc[:324,'Tổng']/data.loc[325,'Tổng'] #Lấy số lượng từng nước chia cho tổng -> số thực
total_ = sum_.sum() #Tính lại tổng các % xem có đúng = 1 hay không
data.loc[:324,'%'] = sum_.map('{:.2%}'.format) #Chuyển đổi sang dạng %, ghi vào cột %
data.loc[325,'%'] = '{:.2%}'.format(total_) #Chuyển đổi sang dạng %, ghi vào TỔNG, %
data.loc[325,'1949':]=data.loc[:324,'1949':].sum(axis=0)

display(data)

tong = data.groupby('Nơi sinh')['Tổng'].sum().sort_values(ascending= False)
tong = tong.iloc[1:] #Bỏ dòng Tổng trên cùng đi
nam2017 = data.groupby('Nơi sinh')['2017'].sum().sort_values(ascending= False)
nam2017 = nam2017.iloc[1:] #Bỏ dòng Tổng trên cùng đi

print('Bảng xếp hạng 10 nước có công dân nhập cư vào NZ cao nhất lịch sử: \n', tong.head(10), '\n')
print('Bảng xếp hạng 10 nước có công dân nhập cư vào NZ cao nhất năm 2017: \n', nam2017.head(10), '\n')
print('Số lượng người nhập cư Việt Nam năm 2017 là:', nam2017['Vietnam'], ' xếp thứ: ', nam2017.index.get_loc('Vietnam'))
print('Số lượng người nhập cư Việt Nam tổng cộng là:', tong['Vietnam'], ' xếp thứ: ', tong.index.get_loc('Vietnam'))
```
